```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv('Iris.csv')
numeric_features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
df[numeric_features].hist(figsize=(12,10))
```

```python
for feature in numeric_features:
    for species in df['Species'].unique():
        sns.histplot(data=df[df['Species'] == species], x=feature, kde=True, label=species)
        
    plt.title(f'Distribution of {feature} by Species')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    
    plt.legend(title='Species')
```

```python
df[numeric_features].boxplot(figsize=(10, 6))
plt.title('Boxplot of Numeric Features')
plt.ylabel('Measurement (cm)')
```
