```python
import pandas as pd

df=pd.read_csv('titanic.csv')
df
```

```python
df.isnull().sum()
```

```python
df=df.drop(['Cabin','Ticket','Name'],axis=1)
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
import seaborn as sns
import matplotlib.pyplot as pt

demography=df['Sex'].value_counts()
pt.pie(demography,labels=demography.index, autopct='%1.1f%%')
pt.title("Demographic Insights")
```

```python
sns.violinplot(df['Age'])
pt.title('Age Distribution')
```

```python
sns.countplot(x='Sex', hue='Survived',data=df)
pt.title('Survival Count by Gender')
```

```python
sns.histplot(data=df, x='Age', hue='Survived',kde=True)
pt.title('Age Distribution of Passengers (Survived vs Not Survived)')
```

```python
sns.countplot(data=df,x='Embarked',hue='Survived')
```

```python
sns.histplot(data=df, x='Fare', bins=30, kde=True, color='skyblue')
pt.title('Distribution of Ticket Prices')
pt.xlabel('Fare')
pt.ylabel('Frequency')
```
