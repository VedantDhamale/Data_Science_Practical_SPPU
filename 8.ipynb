# Distribution Plot
Distribution plot on fare coloumn. purpose of distribution plot: where the actural data is distributed

```python
sns.distplot(df['fare'])
```

```python
# to not to draw line = kde =kernel density estimate
sns.distplot(df['fare'],kde =False)
```

```python
# Group data by passenger class (Pclass)
grouped_data = df.groupby('pclass')

# Create a figure for subplots
fig, axs = plt.subplots(1, 3, figsize=(12, 4))  # Adjust rows/cols and figure size as needed

# Loop through passenge r classes and plot histograms
for i, (pclass, group_df) in enumerate(grouped_data):
    group_df['fare'].hist(ax=axs[i], bins=10, label=f'pclass {pclass}')
    axs[i].set_title(f'Fare Distribution for Pclass {pclass}')
    axs[i].set_xlabel('Fare (USD)')
    axs[i].set_ylabel('Count')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

```python
if 'fare' in df.columns:
    # Group by 'fare' and plot histogram
    df.groupby('fare')['fare'].count().plot(kind='hist', bins=20)
    plt.xlabel('Fare')
    plt.ylabel('Frequency')
    plt.title('Distribution of Ticket Fare')
    plt.show()
else:
    print("The 'fare' column does not exist in the dataset.")
```
