# 9) Data Visualization II
1. Use the inbuilt dataset 'titanic' as used in the above problem. Plot a box plot for distribution of age with respect to each gender along with the information about whether they survived or not. (Column names : 'sex' and 'age').
2. Write observations on the inference from the above statistics.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe(include = 'object')
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['age'] = df['age'].fillna(np.mean(df['age']))
```

```python
df['deck'] = df['deck'].fillna(df['deck'].mode()[0])
```

```python
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode()[0])
```

```python
df.isnull().sum()
```

# Box Plot:

```python
sns.boxplot(df['age'])
plt.show()
```

```python
sns.boxplot(x='sex',y ='age',data=df,hue='survived')
```

```python
sns.boxplot(x='sex',y ='age',data=df)
```

# Violin Plot

```python
sns.violinplot(x ="sex", y ="age", hue ="survived", data = df, split = True)
```

```python
sns.violinplot(x='sex',y ='age',data=df)
```

```python
sns.violinplot(x ="sex", y ="age", hue ="survived", data = df)
```

# Count Plot

```python
sns.countplot(x='survived', data=df)
plt.title("Titanic Survived")
plt.show()
```

# Strip Plot:

```python
sns.stripplot(x='sex',y='age',data=df)
```

```python
sns.stripplot(x= 'sex',y='age',data=df,hue="survived")
```

```python
sns.stripplot(x= 'sex',y='age',data=df,hue="survived",split=True)
```

# Swarm Plot:

```python
sns.swarmplot(x='sex',y ='age',data=df)
```

```python
sns.swarmplot(x='sex',y ='age',data=df,hue = 'survived')
```

```python
sns.swarmplot(x='sex',y ='age',data=df,hue='survived',split=True)
```

```python
sns.boxplot(df['sex'], df["age"], df["survived"]).set_title('Plot for distribution of age with respect to each gender along with the information about whether they survived or not')
plt.show()
```
